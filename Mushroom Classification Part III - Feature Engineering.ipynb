# Introduction
In Part II, we performed simple exploratory data analysis to investigate patterns in the data. 

However, our columns contained nothing but strings. We will need to prepare the data in a way that is suitable for training by dummifying the string values.

### Step 1: Import pandas
Start by importing pandas to read our CSV from Part I.

```python
# Step 1: Import pandas as pd

import pandas as pd
```

### Step 2: Read the CSV from Part I as DataFrame
We will now read the data that we exported from Part I.



```python
# Step 2: Read our cleaned csv from Part I

df = pd.read_csv('Mushroom_DF.csv', index_col= 0)
df
```

### Step 3: Turn 'class' values into binary values
Before we do anything, let's turn the values in 'class' column into binary form, since these are going to be our labels.

```python
# Step 3: Turn the values in 'class' into binaries
# using LabelEncoder from sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder

encode_label = LabelEncoder()
df['class'] = encode_label.fit_transform(df['class'].values)
```

### Step 4: Dummify the rest of the DataFrame
Having transformed the 'class' column, we can now transform the rest of our features.

We will be using pandas get_dummies method to one-hot encode the columns, along with dropping the first column.

```python
# Step 4: Dummify your DataFrame
dummified_df = pd.get_dummies(df, drop_first = True)
dummified_df
```

### Step 6: Export dummy DataFrame 
Export the dummified dataframe into a new csv

```python
# Step 6: Export DataFrame into a CSV

dummified_df.to_csv('dummified_df.csv')
```
